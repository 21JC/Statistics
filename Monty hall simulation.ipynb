```python
import numpy as np
import random
import pandas as pd
```

```python
switch_trial = 0
non_switch_trial = 0
switch_success = 0
non_switch_success = 0
for i in range (0,10000):
    set_up =np.zeros(3)
    prize_index=random.randint(0, 2)
    set_up[prize_index]=1
    set_up_matrix =  pd.DataFrame(np.array([set_up ]))
    initial_guess = random.randint(0, 2)
    narrowed = set_up_matrix.drop([prize_index,initial_guess],axis=1)
    reveal = narrowed.columns[0]
    switch = random.randint(0, 1)
    print("prize is at" + str(prize_index))
    print("initial guess" + str(initial_guess))
    if switch == 0: # didn't switch
        print("didn't switch")
        non_switch_trial+=1
        if initial_guess == prize_index:
            non_switch_success+=1
            print("success")
        else:
            print("failed")
    elif switch ==1:
        print("did switch")
        switch_trial +=1
        new_choice = set_up_matrix.drop([reveal,initial_guess],axis=1).columns[0]
        if new_choice == prize_index:
            switch_success+=1
            print("success")
        else:
            print("failed")
```

```python
switch_success / switch_trial
```

```python
non_switch_success/non_switch_trial
```
